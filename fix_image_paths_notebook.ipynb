# Correction des chemins d'images dans produits.json
Ce notebook charge le JSON, met à jour les chemins d'images et réécrit le fichier avec les nouveaux chemins.

```python
import json
import os
```

```python
root = r'c:\Users\fm005\OneDrive\Desktop\cosmétique\cosmetique'
json_path = os.path.join(root, 'lib', 'data', 'produits.json')
with open(json_path, 'r', encoding='utf-8') as f:
    data = json.load(f)
data[:1]
```

```python
mapping = {
    'crem': 'crem',
    'care': 'care',
    'face': 'face',
    'gom': 'gom',
    'hair': 'hair',
    'soap': 'soap',
    'vit': 'vita',
    'oth': 'others',
}

def corrected_pic(pic):
    if not isinstance(pic, str):
        return pic
    for old_prefix, folder in mapping.items():
        prefix = f'/{old_prefix} ('
        if pic.startswith(prefix) and pic.endswith(').png'):
            number = pic[len(prefix):-5]
            return f'/{folder}/{old_prefix} ({number}).png'
    return pic
```

```python
updated = 0
for category in data:
    for product in category.get('products', []):
        old_pic = product.get('pic')
        new_pic = corrected_pic(old_pic)
        if old_pic != new_pic:
            product['pic'] = new_pic
            updated += 1

with open(json_path, 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

updated
```
